```python
# Read the JSONL file
allArtworks = []
with open(r"api-data\getting-started\allArtworks.jsonl", "r") as rows:
    for row in rows:
        allArtworks.append(json.loads(row))

allArtworks
```

```python
# Create a DataFrame from the list of JSON objects
allArtworks_df = pd.DataFrame(allArtworks)
allArtworks_df
```
