```python
def get_data():
    return pd.read_csv(r"C:\Users\sarit\OneDrive\Desktop\Class_Folder\mygithub\project-3\allMuseums_commaSep.csv")

source_data = get_data()
source_data.head()
```

```python
artist_name = "Henri Matisse"  # Replace Artist_Name with the specific artist you want to filter on
museum_name = "Carnegie Museum of Art"  # Replace Museum_Name with the specific museum name you want to filter on

# Filter the data based on the artist name and museum name
filtered_data = source_data.loc[(source_data['artist'] == artist_name) & (source_data['museum'] == museum_name)]
filtered_data
```

```python
# Create a text input widget
text_input = pn.widgets.TextInput(name='Enter Artist Name', value='', css_classes=['my-text-input'])

# Create a button widget
button = pn.widgets.Button(name='Filter Data', button_type='primary', css_classes=['my-button'])

# Create a DataFrame widget to display the filtered data
filtered_data_widget = pn.widgets.DataFrame(css_classes=['my-dataframe'])

# Apply custom CSS styles to the DataFrame widget
#filtered_data_widget.style.css({'color': 'black', 'background': 'lightgrey'})

# Define a function to handle button click
def filter_data(event):
    artist_name = text_input.value
    #artist_name = "Henri Matisse"
    #print(artist_name)
    filtered_data = source_data.loc[source_data['artist'] == artist_name]
    
    # Display the filtered data or perform any other action
    filtered_data_widget.value = filtered_data
    
button.on_click(filter_data)

# Define custom CSS rules for the widgets
css = '''
.my-text-input {
    background-color: green;
    color: white;
}

.my-button {
    background-color: green;
    color: white;
}

.my-dataframe {
    font-size: 14px;
    border: 1px solid white;
}
'''

# Apply the custom CSS rules to the Panel dashboard
pn.extension(raw_css=[css])
# Create a Panel layout with the widgets
widgets_panel = pn.Column(text_input, button, filtered_data_widget)

# Make the widgets panel servable
widgets_panel.servable()
```
